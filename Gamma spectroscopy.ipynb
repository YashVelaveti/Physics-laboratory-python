## Task 2: Energy calibration of the NaI(Tl) and HPGe detectors <a name="calibration"></a>

```python
### your code goes here 
Table__Cs137_NaI = np.asarray([662])           # -> TODO <- Insert energy [keV] of the first peak
Table_lead_Cs137_NaI = np.asarray([75])           # -> TODO <- Insert energy [keV] of the second peak
Table_1_Na22_NaI = np.asarray([511])           # -> TODO <- Insert energy [keV] of the third peak
Table_2_Na22_NaI = np.asarray([1274.53])           # -> TODO <- Insert energy [keV] of the fourth peak
Table_1_Co60_NaI = np.asarray([1173.2])           # -> TODO <- Insert energy [keV] of the fifth peak
Table_2_Co60_NaI = np.asarray([1332.5])           # -> TODO <- Insert energy [keV] of the fifth peak



#### NO NEED TO EDIT ####
Table_E = np.concatenate([Table__Cs137_NaI,Table_lead_Cs137_NaI,Table_1_Na22_NaI,Table_2_Na22_NaI,Table_1_Co60_NaI,Table_2_Co60_NaI]) #put all energies into one container 'Table_E' for ease of use
mu_peak1 = np.asarray([peak_Cs137_NaI.mu]) #save fitted centroid value
mu_peak2 = np.asarray([peaklead_Cs137_NaI.mu]) #save fitted centroid value
mu_peak3 = np.asarray([peak1_Na22_NaI.mu]) #save fitted centroid value
mu_peak4 = np.asarray([peak2_Na22_NaI.mu]) #save fitted centroid value
mu_peak5 = np.asarray([peak1_Co60_NaI.mu]) #save fitted centroid value
mu_peak6 = np.asarray([peak2_Co60_NaI.mu]) #save fitted centroid value

mu_calibration = np.concatenate([mu_peak1,mu_peak2,mu_peak3,mu_peak4,mu_peak5, mu_peak6]) #put all fitted centroid values in one container 'mu_calibration' for ease of use

print(f'gamma energies [MeV]: {Table_E}')
print(f'Fitted centroid values [ADC]: {mu_calibration}')
#########################

# Same as before, the linear function needs intial guess for the two paramters:
k_guess = 2 # -> intial guess for the slope of the curve
m_guess = 1 # -> initial guess for the intersection of the y-axis


#### NO NEED TO EDIT ####
# Perform the fit
estimates_NaI, covar_matrix = curve_fit(
    fittingFunctions.LineFunc,
    mu_calibration,
    Table_E,
    p0=[k_guess, m_guess]
)

k = estimates_NaI[0]
m = estimates_NaI[1]

# Extract standard deviations (errors) from the covariance matrix
k_err = np.sqrt(covar_matrix[0, 0])
m_err = np.sqrt(covar_matrix[1, 1])

# Plot
plt.figure()
plt.plot(mu_calibration, fittingFunctions.LineFunc(mu_calibration, k, m), color='purple', lw=2, label='linear fit')
plt.scatter(mu_calibration, Table_E, s=100, c='black', marker='*', label='data points')
plt.legend(loc='upper left')
plt.xlabel("Channels")
plt.ylabel("Energy, keV")  # Changed from MeV if your Table_E values are in keV
plt.title("Energy Calibration Curve")
plt.grid(True)
plt.show()

# Print results
print(f"Estimated parameters:")
print(f"  k = {k:.5f}  {k_err:.5f}")
print(f"  m = {m:.5f}  {m_err:.5f}")
#########################

#After executing this cell (Shift + Enter) the fitted parameters can be seen below the plot.
Co60_NaI.calibrate(k,m) #calibrate the Th-228 spectra
Cs137_NaI.calibrate(k,m)
Na22_NaI.calibrate(k,m)
Rock_NaI.calibrate(k,m)

plt.figure(figsize=(10, 6))
# with the data read in with the first routine
plt.step(Co60_NaI.energy, Co60_NaI.counts, where='mid', label='step')
plt.title("Calibrated Co60_NaI spectrum measured by NaI detector") # set title of the plot
plt.xlabel("Energy, keV")     # set label for x-axis 
plt.ylabel("Counts")       # set label for y-axis 
plt.show()

plt.figure(figsize=(10, 6))
# with the data read in with the first routine
plt.step(Cs137_NaI.energy, Cs137_NaI.counts, where='mid', label='step')
plt.title("Calibrated Cs137_NaI spectrum measured by NaI detector") # set title of the plot
plt.xlabel("Energy, keV")     # set label for x-axis 
plt.ylabel("Counts")       # set label for y-axis 
plt.show()

plt.figure(figsize=(10, 6))
# with the data read in with the first routine
plt.step(Na22_NaI.energy, Na22_NaI.counts, where='mid', label='step')
plt.title("Calibrated Na22_NaI spectrum measured by NaI detector") # set title of the plot
plt.xlabel("Energy, keV")     # set label for x-axis 
plt.ylabel("Counts")       # set label for y-axis 
plt.show()

plt.figure(figsize=(10, 6))
# with the data read in with the first routine
plt.step(Rock_NaI.energy, Rock_NaI.counts, where='mid', label='step')
plt.title("Calibrated Rock_NaI spectrum measured by NaI detector") # set title of the plot
plt.xlabel("Energy, keV")     # set label for x-axis 
plt.ylabel("Counts")       # set label for y-axis 
plt.show()
'''
Rock_NaI_energy1 = fittingFunctions.perform_Gaussian_fit(x=Rock_NaI.energy, # your x data 
                                              y=Rock_NaI.counts,           # your y data 
                                              region_start=2000,   # -> TODO <- Insert energy [MeV] from where to START fitting.
                                              region_stop=2500,    # -> TODO <- Insert energy [MeV] from where to STOP fitting.
                                              mu_guess=2250,       # -> TODO <- Insert intial guess of peak centroid (mean value of the peak)
                                              A_guess=100,        # -> TODO <- Insert intial guess for the constant 'A' (amplitude of the peak)
                                              sigma_guess=3)    # -> TODO <- Insert intial guess for the standard diviation (width of the peak)

#After executing this cell (Shift + Enter) the fitted parameters can be seen below the plot.

#bins = 100 # bins per unit of x axis'''
```
